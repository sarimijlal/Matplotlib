## Matplotlib Bars Graph

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (3, 3)

x = np.array(["A","B","C","D"])
y = np.array([10,30,40,20])

plt.bar(x,y)
plt.show()
```

```python
#you can also display the graph horizontally 
plt.barh(x,y,)
plt.show()
```

### Bar width and color

```python
#bar() takes arguments "color" for the color of bars and "width" for setting the bar width
plt.bar(x,y, color = 'CornflowerBlue', width = 0.6) #default width is 0.8
plt.show()
```

```python
#for horizontal bars we use height argument insted of width
plt.barh(x,y, height = 0.5)
# plt.bar(x,y, height = 0.5) this will throw and error

plt.show()
```

### Labels 

```python
#adding x,y and general label and properties of labels is pretty much the same as other plots.

#suppose we have a bar plot representing the usage of different programming languages with repect to its users.
x1= np.array(["Java", "Python", "JavaScript", "C", "C++"])
y1= np.array([40, 80, 60, 35, 45])

plt.figure(figsize= (4,4))

#setting color of each bar
c = ['#A9A9A9', '#696969','#808080', '#D3D3D3', '#C0C0C0' ]

plt.bar(x1,y1, color = c , width = 0.6, label = "trend") #default width is 0.8

plt.xlabel("Languages", fontsize = 12, color = "CornflowerBlue")
plt.ylabel("Users", fontsize = 12, color = "CornflowerBlue")
plt.title ("Sarim",fontsize = 15, color = "CornflowerBlue")

plt.xticks(rotation = 10)
plt.legend()
plt.show()
```

### Multiple bars in one graph

```python
x1= np.array(["Java", "Python", "JS", "C", "C++"])
y1= np.array([40, 80, 60, 35, 45])
z1= np.array([20, 60, 40, 20, 35])

plt.figure(figsize= (6,5))

width = 0.4
p= np.arange(len(x1))
p1= [j+width for j in p]

#setting color of each bar
c = ['#A9A9A9', '#696969','#808080', '#D3D3D3', '#C0C0C0' ]

plt.bar(p,y1, width, color = c, edgecolor = 'black', label = "After AI") #default width is 0.8
plt.bar(p1,z1, width, color= c, edgecolor= 'black', label = "Before AI")
plt.xlabel("Languages", fontsize = 10)
plt.ylabel("Users", fontsize = 10)
plt.title ("Sarim")

plt.xticks(p+width/2, x1)
plt.legend()
plt.show()
```
