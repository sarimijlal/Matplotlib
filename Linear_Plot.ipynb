# Linear Plot

```python
import matplotlib.pyplot as plt
import numpy as np
#plt.rcParams["figure.figsize"] = (3, 3)  # Width, Height in inches
x= [1,2,3,4]
y= [2,4,6,8]
plt.plot(x,y)
plt.show
```

### Passing Numpy arrays as input vectors

```python
#passing numpy arrays as input vectors
x_axis = np.array([2,8])
y_axis = np.array([3,10])
plt.plot(x_axis,y_axis,'o') #ploting only end points
plt.show
```

### Ploting multiple points

```python
#ploting multiple points starting from (1,3) to (2,8) then to (6,1) and finally (8,10)
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker= 'o',)
plt.show()
```

### Ploting only 1 axis

```python
yaxis = np.array([1,2,3,4])

plt.plot(yaxis)
plt.show()
```

```python
xaxes = np.array([3,7,9])
plt.plot(xaxes)
plt.show()
```

### Changing line appearance 

```python
#using the previous code

#ploting multiple points starting from (1,3) to (2,8) then to (6,1) and finally (8,10)
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'o:g') 
# or adding another argument "linestyle"
# plt.plot(xpoints, ypoints, linestyle= 'dotted' or ls=':')
plt.show()
```

### Adding Title, labels and grid 

```python
#ploting multiple points starting from (1,3) to (2,8) then to (6,1) and finally (8,10)
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker= 'o')
plt.title("Title-here")
#to add labels you can use functions "xlabel()" and "ylabel()"
plt.xlabel("anyxlabel")
plt.ylabel("anyylabel")
plt.grid() #you can specify the gird line by passing the 'axis' parameter 
#plt.grid(axis = "y", c = "g", lw = 0.5)
plt.show()

#you can also change the position and font of title and label using some parameters.
```

## Sub Plots

```python
#with the usage of 'subplot()' function we can add multiple plots in a single figure
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (7, 3)  # Width, Height in inches

#plot 1
x= np.array([0,1,2,3])
y= np.array([3,8,1,10])

plt.subplot(1,2,1) 
plt.plot(x,y)
plt.title('ind tit-1')

#plot 2
x1= np.array([0,1,2,3])
y1= np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x1,y1)
plt.title('ind tit-2')

plt.suptitle("2-plots") #adds the title of whole figure
plt.show()

#the 'subplot()' function takes 3 arguments such as in our case (1,2,1), the first argument is the no of rows which figure contains and the second argument is the number of coloumns and the third argument is the no of plot
#you can make as much plot as you want, all you need to do is to adjust the rows, coloumn and plot number accordingly
```
