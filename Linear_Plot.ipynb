# Linear Plot

```python
import matplotlib.pyplot as plt
import numpy as np
#plt.rcParams["figure.figsize"] = (3, 3)  # Width, Height in inches
x= [1,2,3,4]
y= [2,4,6,8]
plt.plot(x,y)
plt.show
```

### Passing Numpy arrays as input vectors

```python
#passing numpy arrays as input vectors
x_axis = np.array([2,8])
y_axis = np.array([3,10])
plt.plot(x_axis,y_axis,'o') #ploting only end points
plt.show
```

### Ploting multiple points

```python
#ploting multiple points starting from (1,3) to (2,8) then to (6,1) and finally (8,10)
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

### Ploting only 1 axis

```python
yaxis = np.array([1,2,3,4])

plt.plot(yaxis)
plt.show()
```
